```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r'/content/loan.csv')
dataset.head(20)
```

```python
dataset.isnull().sum()
```

```python
dataset.fillna(10) # This will fill the NaN value with 10
```

## However, this method would insert the value 10 into the columns "Self_Employed" and "Property_Area", which contain categorical data. Since these columns are not numerical, filling them with a numeric value like 10 would not be meaningful.

**Numerical columns → can fill with mean / median / numbers**

**Categorical columns → can fill in 3 ways:<br>
                      i) Backward Fill<br>
                      ii)Forward Fill<br>
                      iii) Fill with Mode**






**In data preprocessing:**
---------------------

Forward Fill (ffill) → fills with the previous value

Backward Fill (bfill) → fills with the next value

# **Finding catagorical type data**

```python
dataset.info()
```

# **Backward Fill**


```python
dataset = dataset.bfill()  # or dataset.bfill(inplace = True)
dataset.head(4)  # See the NaN value is replaced with 128.0 in LoanAmount column
```

#**Forward Fill**

```python
dataset.fillna(axis = 1)
```

## As `LoanAmount` has `NaN` in the first row, it cannot be filled with the previous data in the same column.  
## That's why `axis=1` is used, which fills it with the previous column's data in the same row.

## **For using previous columns's data: <br>**
 dataset.bfill(axis = 1)<br>
 dataset.ffill(axis = 1)

# **💡 In short:<br>**

###axis=0 → column-wise filling (up & down)<br>
###axis=1 → row-wise filling (left & right)

# **Mode Fill**

```python
dataset["Gender"].mode()
```

```python
dataset.fillna({"Gender": dataset["Gender"].mode()[0]}, inplace = True) # dataset.fillna({col, value}, inplace = True)
```

To Fill All the NaN values of Object type data with mode
