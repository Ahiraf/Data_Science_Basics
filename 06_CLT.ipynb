# ** CLT: How the sampling distribution approaches a normal distribution.**

## According to the Central Limit Theorem, the sampling distribution of the sample mean becomes approximately normal as the sample size increases.

### Even if the population is not normally distributed, the distribution of the sample mean will become approximately normal as the sample size increases.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
pop_data = [np.random.randint(10, 100) for i in range(10000)]# Random number will come every time whenener running, [pop_data = population data]

pop_table = pd.DataFrame({"Pop_data": pop_data})
````

````python
pop_table
````

````python
plt.figure(figsize = (4, 3))
sns.kdeplot(x = "Pop_data", data = pop_table)
plt.show()
````



```
# Now every sample group should be less than 10% of the population. So,
(10000*10)/100 = 1000
# And every sample group size should be greater than 30
```



````python
sample_mean = []
for no_sample in range(60):
  sample_data = []
  for data in range(500):
    sample_data.append(np.random.choice(pop_data))  # Every sample group will store in sample_data list.
  sample_mean.append(np.mean(sample_data)) # Every mean of eache sample_data group will save in sample_mean

sample_mean
````

````python
sample_M = pd.DataFrame({"Sample_Mean": sample_mean})
sns.kdeplot(x = "Sample_Mean", data = sample_M)
plt.show()
````
