```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({"Size": ["s", "m", "l", "xl", "s", "m", "l", "s", "s", "l", "xl", "m"]})
df.head(3)
```

```python
ord_data = [["s", "m", "l", "xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
# If categories='auto', the encoder assigns numbers based on the alphabetical order of categories.
# Here we provide ord_data to define the desired order of categories manually.
```

```python
oe.fit(df[["Size"]])
```

```python
df["Size_en"] = oe.transform(df[["Size"]])
df
```

# **Encoding using Map Function**

```python
ord_data1 = {"s": 0, "m": 1, "l": 2, "xl": 3}
# Custom mapping for ordinal categories; any numerical values can be assigned using map(),
# e.g., 5, 6, 7, 8 instead of 0, 1, 2, 3.
```

```python
df["Size_en1"] = df["Size"].map(ord_data1)
df
```
