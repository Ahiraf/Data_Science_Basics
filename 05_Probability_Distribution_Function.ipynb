````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
dataset = pd.read_csv('/content/tips.csv')
````

````python
dataset.head(3)
````

````python
dataset = dataset.drop(columns=['CC Number']) #That is a credit card number,which should not be used in statistical analysis because:

#It’s just an ID
#It has no mathematical meaning
#It will create meaningless huge values

dataset.head(3)
````

````python
dataset.isnull().sum()
````

````python
dataset.info() # for checking the datatypes of each column
````

````python
data_corr = dataset.select_dtypes(["float64", "int64"]).corr()
data_corr
````



```
# So from this it can be seen that all the corr is between -1 to 1 :
total_bill and total_bill correlation = 1
total_bill and tip correlation = 0.675734 (strong corr)
total_bill and size correlation = 0.598315(strong corr)
thus all the corr can be shown....
```



````python
data_cov = dataset.select_dtypes(["float64", "int64"]).cov()
data_cov
````



```
# It can be seen that all the values of covariance is from minus intinity to infinity..
```



````python
plt.figure(figsize = (4, 3)) # to make the map smaller
sns.heatmap(data_corr, annot = True)
plt.show()
````

````python
#plt.figure(figsize = (4, 3))
sns.heatmap(data_cov, annot = True)
plt.show()
````
