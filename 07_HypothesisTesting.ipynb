```python
import scipy.stats as st
import numpy as np
```

```python
z_table = st.norm.ppf(1-0.05)  # ppf() accept the value of  1 - alpha or level of significance. then it automatically gives the z value as output.
z_table
```

# **Finding Zcal**

```python
s_x = 90
n = 81
p_u = 82
p_std = 20
```

```python
z_cal = (s_x - p_u)/ (p_std / np.sqrt(n))
z_cal
```

```python
if z_table < z_cal:
  print("ha is right")
else:
  print("ho is right")
```
