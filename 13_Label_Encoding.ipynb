```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({"name": ["cow", "dog", "cat", "tiger", "buffalo"]})
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit_transform(df[["name"]]) # Returns an array of encoded label values
```

```python
df["en_name"] = le.fit_transform(df[["name"]])
df
```

```python
dataset = pd.read_csv('/content/loan.csv')
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
la = LabelEncoder()
la.fit(dataset["Property_Area"])
```

```python
la.transform(dataset["Property_Area"])  # The output shows that labels are assigned such as Urban = 2, Rural = 0, etc.
```

## **If this column contains NaN values, they will also appear after label encoding.**
