### **In Machine Learning, encoding means converting categorical (text) data into numerical form so that a machine learning model can understand it.**

### **Most ML algorithms work with numbers, not text.**
### **So we transform values like "Male", "Female", "Urban", "Rural" into numbers. This transformation process is called encoding.**

## **Nominal data (no order) → One-Hot Encoding**<br>**Ordinal data (has order) → Label / Ordinal Encoding**


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/content/loan.csv')
dataset.head(3)
```

```python
dataset.isnull().sum()
```

### **Gender, Married, Education, Property_Area columns are nominal categorical variables, so we use One-Hot Encoding.**

### **Among these columns Gender and Married columns have NaN values.**

```python
dataset.fillna({"Gender": dataset["Gender"].mode()[0]}, inplace = True)
dataset.fillna({"Married": dataset["Married"].mode()[0]}, inplace = True)
```

```python
dataset.isnull().sum()
```

# **# One-Hot Encoding can be performed in two ways: using Pandas or using Scikit-learn.**

# **Using Pandas:**

```python
en_data = dataset[["Gender", "Married"]]

pd.get_dummies(en_data)
```

Here True = 1 and False = 0 or boolean types data.

```python
pd.get_dummies(en_data).info()
# After encoding, the "Gender" column will be converted into two columns: Gender_Female and Gender_Male.
# Similarly, the "Married" column will be converted into Married_No and Married_Yes.
```

# **For Example:**
| Gender | Gender_Female | Gender_Male |
| ------ | ------------- | ----------- |
| Female | 1             | 0           |
| Male   | 0             | 1           |


# **Using Scikit-learn:**

```python
!pip install sklearn.preprocessing
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
ohe.fit_transform(en_data) # fit_transform() converts categorical data into numerical form using a Scikit-learn encoder
```

### **Small concept clarification (important)**
 **fit_transform() actually does two steps:**

###**1️⃣ fit()**
**Learns the categories from the data.<br>
Example: Male, Female**

###**2️⃣ transform()**
**Converts them into numbers or encoded vectors**


## ---

**So, fit_transform() = fit() + transform()**

**Sparse matrix: a matrix in which most elements are 0 and only a few are non-zero.**

**This happens often after One-Hot Encoding, which is why libraries like Scikit-learn return sparse matrices to save memory.**

```python
# Converting Sparse Matrix to Array

arr = ohe.fit_transform(en_data).toarray()
arr
```

```python
#Converting array to DataFrame

pd.DataFrame(arr, columns = ["Gender_Female", "Gender_Male", "Married_No", "Married_Yes"])
```

# **One-Hot Encoding creates multiple new columns, so we drop one column to avoid redundancy.**

**If we already know Gender_Female, we automatically know Gender_Male.**
**Example:**

### **Female = 1 → Male = 0** <br> **Female = 0 → Male = 1**

**So one column is redundant.**

```python
ohe = OneHotEncoder(drop = "first") # pd.get_dummies(dataset, drop_first=True)
ohe.fit_transform(en_data)

arr = ohe.fit_transform(en_data).toarray()
arr
```

```python
pd.DataFrame(arr, columns = ["Gender_Male", "Married_Yes"])
```

##**After using drop='first', if Gender_Male = 1 then the gender is Male, and if Gender_Male = 0 then the gender is Female.**
