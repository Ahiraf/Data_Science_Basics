# **Measurement of Variability **
## It measures the spreadness of data
### The smaller the spread of the data, the better.

# **Mean Absolute Deviation(MAD)**

```python
sec_A = np.array([75, 65, 73, 68, 72, 67])
sec_B = np.array([90, 47, 43, 96, 93, 51])
no = np.array([1, 2, 3, 4, 5, 6])
```

```python
mn_A = np.mean(sec_A)
mn_B = np.mean(sec_B)
(mn_A, mn_B)
```

```python
plt.figure(figsize = (10, 4))
plt.scatter(sec_A, no, label = "sec_A") # plt.scatter(x, y)
plt.scatter(sec_B, no, color = "red", label = "sec_B")
plt.plot([mn_A for i in range(0, 8)], [i for i in range(0, 8)], color = "blue")
plt.legend()
plt.show()
```

IMG_0402.jpeg

```python
mad_A = np.sum(np.abs(sec_A - mn_A)) / len(sec_A)
mad_B = np.sum(np.abs(sec_B - mn_B)) / len(sec_B)
print(mad_A)
print(mad_B)
# so A section has less deviation from center. So this section should be selected for the competition.
```

# **Standard Deviation**
## The smaller the SAD the better

```python
stdA = np.std(sec_A)
stdB = np.std(sec_B)
print(stdA)
print(stdB)
```

# **Variance**
### variance = square of std

```python
var_A = np.var(sec_A)
var_B = np.var(sec_B)
(var_A, var_B)
```
